```python
import time
import traceback
import pandas as pd
from argparse import ArgumentParser
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
import os
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException, StaleElementReferenceException, WebDriverException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

EXTENSION_PATH = r"C:/Users/Ali Hassan/AppData/Local/Google/Chrome/User Data/Profile 7/Extensions/ifibfemgeogfhoebkmokieepdoobkbpo/3.7.2_0"

options = uc.ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--start-maximized")
options.add_argument(f"--load-extension={EXTENSION_PATH}")

driver = uc.Chrome(options=options)

driver.set_page_load_timeout(120)
driver.set_script_timeout(120)

driver.set_window_size(1280, 800)

time.sleep(5)
driver.refresh()

BASE_URL = "https://www.cyberbackgroundchecks.com"

def read_excel(file_path):
    """Read the Excel file and return required columns."""
    try:
        df = pd.read_excel(file_path)
        required_columns = ["City", "State", "Owner 1 First Name", "Owner 1 Last Name"]
        df = df[required_columns]
        return df
    except Exception as e:
        print(f"Error reading Excel file: {e}")
        traceback.print_exc()
        return None

def wait_for_page_load(timeout=60):
    """Wait for the page to be fully loaded before continuing."""
    try:
        WebDriverWait(driver, timeout).until(
            lambda d: d.execute_script('return document.readyState') == 'complete'
        )
        print("Page fully loaded and ready.")
    except TimeoutException:
        print("Error: Page did not load in time.")
        # Raise an exception, or optionally just log and continue
        raise

def wait_for_captcha_and_solve():
    """Wait for the CAPTCHA (Cloudflare Turnstile) to appear and be solved."""
    try:
        WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div#cf-turnstile"))
        )
        print("CAPTCHA detected. Waiting for it to be solved...")

        WebDriverWait(driver, 60).until(
            EC.invisibility_of_element_located((By.CSS_SELECTOR, "div#cf-turnstile"))
        )
        print("CAPTCHA solved. Proceeding with next steps.")
    except TimeoutException:
        print("No CAPTCHA detected or it took too long to solve.")

def search_for_phone_numbers(first_name, last_name, address):
    """
    Search the website for phone numbers based on the provided name and address.
    Returns:
        List of phone numbers (if found) or None (if none found).
    """
    try:
        driver.get(BASE_URL)
        time.sleep(2)
        driver.refresh()
        time.sleep(1)

        wait_for_captcha_and_solve()
        wait_for_page_load()

        WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.ID, "SearchCriteriaViewModel_FirstName"))
        )
        first_name_input = driver.find_element(By.ID, "SearchCriteriaViewModel_FirstName")
        last_name_input = driver.find_element(By.ID, "SearchCriteriaViewModel_LastName")
        address_input = driver.find_element(By.ID, "SearchByName_AddressLine2")
        search_button = driver.find_element(By.ID, "button-search-by-name")

        first_name_input.clear()
        first_name_input.send_keys(first_name)
        last_name_input.clear()
        last_name_input.send_keys(last_name)
        address_input.clear()
        address_input.send_keys(address)

        search_button.click()

        wait_for_page_load()
        wait_for_captcha_and_solve()

        WebDriverWait(driver, 60).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a.btn.btn-primary.btn-block'))
        )
        links = driver.find_elements(By.CSS_SELECTOR, 'a.btn.btn-primary.btn-block')

        match_found = False
        for link in links:
            title = link.get_attribute('title')
            if address.lower() in title.lower():
                link.click()
                match_found = True
                break  # Click only the first match

        wait_for_page_load()
        wait_for_captcha_and_solve()

        if not match_found:
            print(f"No matching results found for {first_name} {last_name}.")
            return None

        phone_section = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".col-md-12.text-secondary"))
        )

        none_found = False
        if none_found:
            return None

        phone_links = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a.phone"))
        )
        phone_numbers = [link.text for link in phone_links]
        return phone_numbers if phone_numbers else None

    except TimeoutException:
        print(f"Timeout: No results found or page load timed out for {first_name} {last_name}.")
        return None
    except StaleElementReferenceException:
        print(f"Stale element error while searching for {first_name} {last_name}.")
        return None
    except WebDriverException as wd_err:
        print(f"WebDriver error for {first_name} {last_name}: {wd_err}")
        return None
    except Exception as e:
        print(f"Unknown error during search for {first_name} {last_name}: {e}")
        traceback.print_exc()
        return None
    finally:
        pass

def save_to_excel(data, output_file):
    """Save the extracted data to an Excel file."""
    try:
        df = pd.DataFrame(data)
        df.to_excel(output_file, index=False)
        print(f"Results saved to {output_file}")
    except Exception as e:
        print(f"Error saving to Excel: {e}")
        traceback.print_exc()

def main(input_file, output_file):
    """Main function to execute the automation process."""
    data = read_excel(input_file)
    if data is None:
        print("Failed to load data. Exiting.")
        return

    results = []
    row_count = len(data.index)

    for index, row in data.iterrows():
        first_name = str(row["Owner 1 First Name"])
        last_name = str(row["Owner 1 Last Name"])
        city = str(row["City"])
        state = str(row["State"])
        address = f"{city}, {state}"

        print(f"\nProcessing Row {index+1}/{row_count}: {first_name} {last_name} from {address}...")

        try:
            phone_numbers = search_for_phone_numbers(first_name, last_name, address)

            if phone_numbers:
                phone_numbers_str = ", ".join(phone_numbers)
                results.append({
                    "First Name": first_name,
                    "Last Name": last_name,
                    "Address": address,
                    "Phone Numbers": phone_numbers_str
                })
            else:
                results.append({
                    "First Name": first_name,
                    "Last Name": last_name,
                    "Address": address,
                    "Phone Numbers": "Not found"
                })

        except Exception as e:
            # Catch any unexpected error that might break the loop
            print(f"**Error in row {index+1} ({first_name} {last_name}):** {e}")
            traceback.print_exc()
            # Add a placeholder indicating an error occurred
            results.append({
                "First Name": first_name,
                "Last Name": last_name,
                "Address": address,
                "Phone Numbers": "ERROR"
            })
        finally:
            try:
                driver.get(BASE_URL)
                time.sleep(2)
            except Exception as nav_err:
                print(f"Error returning to homepage: {nav_err}")
                # Even if we fail here, just continue.

    save_to_excel(results, output_file)

if __name__ == "__main__":
    input_file = "Property Export Pre+foreclosure+atlanta.xlsx"
    output_file = "phone_numbers.xlsx"

    try:
        main(input_file, output_file)
    except Exception as e:
        print(f"Fatal error in main: {e}")
        traceback.print_exc()
    finally:
        # Attempt to close the driver gracefully
        try:
            driver.quit()
        except:
            pass
```
